```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')

# Sample customer reviews
reviews = [
    "The store has great products and excellent customer service.",
    "I'm really disappointed with the quality of their products.",
    "The staff is friendly and helpful.",
    "Terrible experience. I would not recommend this store to anyone.",
    "The prices are reasonable and the selection is good."
]

# Preprocessing and tokenization
stop_words = set(stopwords.words('english'))

tokens = []
for review in reviews:
    review = review.lower()
    words = word_tokenize(review)
    words = [word for word in words if word.isalnum() 
              and word not in stop_words]
    tokens.extend(words)

# WordCloud generation
wordcloud = WordCloud().generate(' '.join(tokens))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Sentiment analysis using TextBlob
polarity = 0
subjectivity = 0

for review in reviews:
    blob = TextBlob(review)
    polarity += blob.sentiment.polarity
    subjectivity += blob.sentiment.subjectivity

average_polarity = polarity / len(reviews)
average_subjectivity = subjectivity / len(reviews)

print("Average Polarity:", average_polarity)
print("Average Subjectivity:", average_subjectivity)

# Sentiment analysis using NLTK Vader
sia = SentimentIntensityAnalyzer()

for review in reviews:
    sentiment = sia.polarity_scores(review)
    print("Review:", review)
    print("Sentiment:", sentiment)
    print()
```
